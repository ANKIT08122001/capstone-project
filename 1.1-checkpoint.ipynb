```python
data1 = pd.read_csv('C:/Users/ankit27/Desktop/data/data_file_1.csv')
data2 = pd.read_csv('C:/Users/ankit27/Desktop/data/data_file_2.csv')
data3 = pd.merge(data1,data2
             )

print(data3)
```

# AGE

```python
x =data3.groupby(['Age','Product_Category_1'],  as_index=False)['Product_ID'].count()
x.head(100)
```

```python
z = x.groupby(['Age'])['Product_ID'].transform(max) == x['Product_ID']
x[z]
```

CONCLUSION:

We observe that Product category 5 is most Purchased so we may provide discount on this category
For Age group 55+ we may provide a discount on Product category 8.
